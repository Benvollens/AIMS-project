## Data collection 

```python
Compounds = list(df['Compounds'])
#Compounds
```

```python
big_list = []
for compound in Compounds:
    sub_elmts = []
    for letter in compound:
        # Check if the character is a letter (upper or lower)
        if letter.isalpha():
            sub_elmts.append(letter)
        else:
            sub_elmts.append(letter)
    big_list.append(sub_elmts[:-1])
print(big_list)
```

```python
main_list = []
all_distincts = []
for sublist in big_list:
    new_sublist=[]
    if sublist[1].lower()==sublist[1]:
        new_sublist.append(sublist[0]+sublist[1])
        if sublist[3].lower()==sublist[3]:
            new_sublist.append(sublist[2]+sublist[3])
            if sublist[-1].lower()==sublist[-1]:
                new_sublist.append(sublist[-2]+sublist[-1])
            else:
                new_sublist.append(sublist[-1])
        else:
            new_sublist.append(sublist[2])
            if sublist[-1].lower()==sublist[-1]:
                new_sublist.append(sublist[-2]+sublist[-1])
            else:
                new_sublist.append(sublist[-1])
            
    else:
        new_sublist.append(sublist[0])
        if sublist[2].lower()==sublist[2]:
            new_sublist.append(sublist[1]+sublist[2])
            if sublist[-1].lower()==sublist[-1]:
                new_sublist.append(sublist[-2]+sublist[-1])
            else:
                new_sublist.append(sublist[-1])
        else:
            new_sublist.append(sublist[1])
            new_sublist.append(sublist[2])
    main_list.append(new_sublist) 
    all_distincts += new_sublist
all_distincts = list(set(all_distincts))
print(main_list) 
```
